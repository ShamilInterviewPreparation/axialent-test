```python
from src.data_preparation import load_dataset
```

```python
dataset_path = "../data/dataset.jsonl"
dataset = load_dataset(dataset_path)
```

```python
from src.analysis.quantative import get_quantative_analysis


analysis_df = get_quantative_analysis(dataset)
analysis_df = analysis_df.iloc[1:]
analysis_df
```

```python
import matplotlib.pyplot as plt

feedback_percentages = [
    len(analysis_df["feedback_Q1"].dropna()) / len(analysis_df) * 100,
    len(analysis_df["feedback_Q2"].dropna()) / len(analysis_df) * 100,
    len(analysis_df["feedback_Q3"].dropna()) / len(analysis_df) * 100,
]

plt.figure(figsize=(5, 3))
plt.bar(["Q1", "Q2", "Q3"], feedback_percentages)
plt.title("Percentage of Feedback Responses by Question")
plt.ylabel("Percentage of Responses")
plt.ylim(0, 100)

# Add percentage labels on top of bars
for i, v in enumerate(feedback_percentages):
    plt.text(i, v + 1, f"{v:.1f}%", ha="center")

plt.show()
```

```python
analysis_df.to_csv("../data/quantative_df.csv", index=False)
```
