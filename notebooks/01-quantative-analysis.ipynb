```python
# drop first row because of large amount of api errors (53)

final_df = analysis_df[analysis_df["conversation_id"] != 0]
final_df
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Filter out rows with empty feedback values
filtered_df = final_df.dropna(subset=["overall_feedback"])

feedback_values = filtered_df["overall_feedback"].unique()
conversation_length_by_feedback = filtered_df.groupby("overall_feedback")["conversation_length"].mean()

# Also calculate mean for rows without feedback
no_feedback_mean = final_df[final_df["overall_feedback"].isna()]["conversation_length"].mean()

plt.figure(figsize=(10, 4))

# Create a new Series with "No Feedback" as the first entry
all_data = pd.Series([no_feedback_mean], index=["No Feedback"])
all_data = pd.concat([all_data, conversation_length_by_feedback])

# Plot all data with "No Feedback" as the first bar
ax = all_data.plot(kind="bar", color=["lightgray"] + ["skyblue"] * len(conversation_length_by_feedback))

# Add value labels on top of bars
for i, v in enumerate(all_data):
    ax.text(i, v + 0.5, f"{v:.1f}", ha="center")

plt.title("Average Conversation Length by Feedback Score (excluding feedback)")
plt.xlabel("Feedback Score")
plt.legend(["No Feedback", "With Feedback"])
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Filter out rows with empty feedback values
filtered_df = final_df.dropna(subset=["overall_feedback"])

feedback_values = filtered_df["overall_feedback"].unique()
conversation_length_by_feedback = filtered_df.groupby("overall_feedback")["user_words"].mean()

# Also calculate mean for rows without feedback
no_feedback_mean = final_df[final_df["overall_feedback"].isna()]["user_words"].mean()

plt.figure(figsize=(10, 6))

# Create a new Series with "No Feedback" as the first entry
all_data = pd.Series([no_feedback_mean], index=["No Feedback"])
all_data = pd.concat([all_data, conversation_length_by_feedback])

# Plot all data with "No Feedback" as the first bar
ax = all_data.plot(kind="bar", color=["lightgray"] + ["skyblue"] * len(conversation_length_by_feedback))

# Add value labels on top of bars
for i, v in enumerate(all_data):
    ax.text(i, v + 0.5, f"{v:.1f}", ha="center")

plt.title("Average User Words by Feedback Score (excluding feedback)")
plt.xlabel("Feedback Score")
plt.legend(["No Feedback", "With Feedback"])
```

## From the following diagram we can not say that there is a strong correlation between the feedback score and the conversation length.
## However, we can see that user rated 5.0 tends to user more words with same amount of messages. 
## Thus, we should encourage users to use more words, explaining their difficult situations - possibly if user do not use more words, ask him more questions or directly ask him to explain his situation more detailed and deeper.
